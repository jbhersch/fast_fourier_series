```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
from ffs import FFS
import seaborn as sns

sns.set_style("whitegrid")

%matplotlib inline

save_plots = False
```

```python
x = np.linspace(-2,2,50)
y = x**3
d1_y = 3*x**2
d2_y = 6*x
fs = FFS(x,y,0)
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs.evaluate(x),'r', label = 'Fourier Series Approximation')
# plt.plot(x, y - fs.evaluate(x), 'g', label='Error')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.title('Fourier Series Representation of the Standard Cubic Function', fontsize=25)
plt.legend(loc = 'upper left', prop={'size':20})

plt.tight_layout()
if save_plots:
    plt.savefig('images/cubic_no_padding.png')
plt.show()
```

```python
xx = np.hstack((x, fs.xb + fs.h))


plt.figure(figsize=(15,10))
plt.plot(xx,fs.evaluate(xx),'b', label="First Period")
plt.plot(xx + fs.L,fs.evaluate(xx + fs.L),'r', label="Second Period")
plt.title("Cubic Function With Two Periods", fontsize=25)
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.xlim((-2.5,6.5))
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc = 'upper left', prop={'size':20})

plt.tight_layout()
# plt.savefig('images/cubic/cubic_two_periods.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,d1_y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs.evaluate(x,deriv=1),'r', label = 'Fourier Series Approximation')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
# plt.ylim((-0.1, 1.1))
plt.title('First Derivative of Standard Cubic Function', fontsize=25)
plt.legend(loc = 'upper center', prop={'size':20})

plt.tight_layout()
# plt.savefig('images/cubic_1st_derivative_no_padding.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,d2_y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs.evaluate(x,deriv=2),'r', label = 'Fourier Series Approximation')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
# plt.ylim((-0.1, 1.1))
plt.title('Second Derivative of Standard Cubic Function', fontsize=25)
plt.legend(prop={'size':20})

plt.tight_layout()
if save_plots:
    plt.savefig('images/cubic_2nd_derivative_no_padding.png')
plt.show()
```

```python
fs_pad = FFS(x,y,50)
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs_pad.evaluate(x),'r', label = 'Fourier Series Approximation')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.title('Cubic Function With Padding', fontsize=25)
plt.legend(loc = 'upper left', prop={'size':20})

plt.tight_layout()
if save_plots:
    plt.savefig('images/cubic_with_padding.png')
plt.show()
```

```python
# xpad = np.hstack((fs_pad.x[fs_pad.n0:], [fs_pad.x[-1] + fs_pad.h]))
xpad = fs_pad.x[fs_pad.n0:]

plt.figure(figsize=(15,10))
plt.plot(x,y,'b', label = 'Input Function')
plt.plot(xpad,fs_pad.evaluate(xpad),'r', label = 'Padding')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.title('Cubic Function Padding Display', fontsize=25)
plt.legend(loc = 'upper left', prop={'size':20})

plt.tight_layout()
# plt.savefig('images/cubic_padding_display.png')
plt.show()
```

```python
xx = np.hstack((fs_pad.x, fs_pad.xb + fs_pad.h))


plt.figure(figsize=(15,10))
plt.plot(x,fs_pad.evaluate(x),'b', label="First Period Input Function", linewidth = 10)
plt.plot(x + fs_pad.L,fs_pad.evaluate(x + fs_pad.L),'r', label="Second Period Input Function", linewidth = 10)
plt.plot(xx,fs_pad.evaluate(xx),'b', label="First Period Padding")
plt.plot(xx + fs_pad.L,fs_pad.evaluate(xx + fs_pad.L),'r', label="Second Period Padding")
plt.title("Padded Cubic With Two Periods", fontsize=25)
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.xlim((-2.5,15))
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc = 'upper center', prop={'size':20}, ncol = 2)

plt.tight_layout()
# plt.savefig('images/cubic_two_periods_with_padding.png')
plt.show()
```

```python
# xx = np.array([fs2.x[fs2.n0 - 2], fs2.x[fs2.n0 - 1], fs2.xb, fs2.xb + fs2.h])
# yy = np.array([fs2.y[fs2.n0 - 2], fs2.y[fs2.n0 - 1], fs2.y[0], fs2.y[1]])

xx = np.array([1,2,6,7])
yy = np.array([3,4,1,2])

x1, x2, x3, x4 = xx
y1, y2, y3, y4 = yy

# linear interpolations
s12_1 = lambda t: ((x2 - t)*y1 + (t - x1)*y2) / (x2 - x1) # between x1 and x2
s23_1 = lambda t: ((x3 - t)*y2 + (t - x2)*y3) / (x3 - x2) # between x2 and x3
s34_1 = lambda t: ((x4 - t)*y3 + (t - x3)*y4) / (x4 - x3) # between x3 and x4

# quadratic interpolations
s13_2 = lambda t: ((x3 - t)*s12_1(t) + (t - x1)*s23_1(t)) / (x3 - x1) # between x1 and x3
s24_2 = lambda t: ((x4 - t)*s23_1(t) + (t - x2)*s34_1(t)) / (x4 - x2) # between x2 and x4

# cubic interpolations
s14_3 = lambda t: ((x4 - t)*s13_2(t) + (t - x1)*s24_2(t)) / (x4 - x1) # between x1 and x4
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

ax.plot(xx, yy, 'o')

ax.plot(np.linspace(x1,x3,50), s13_2(np.linspace(x1,x3,50)), 'r')
ax.plot(np.linspace(x2,x4,50), s24_2(np.linspace(x2,x4,50)), 'm')

ax.plot(np.linspace(x1,x4,100), s14_3(np.linspace(x1,x4,100)), 'g')

ax.plot(xx[:2], yy[:2], 'b')
ax.plot(xx[1:3], yy[1:3], 'b')
ax.plot(xx[2:], yy[2:], 'b')

ax.set_xlim(0, 8)
ax.set_xticks(xx)
ax.set_xticklabels(['x1','x2','x3','x4'], fontsize=15)

# ax.set_ylim(-8.5, 8.5)
ax.set_yticks(yy)
ax.set_yticklabels(['y1','y2','y3','y4'], fontsize=15)

ax.text(1.5, 3.4, '$p_{1,2}^1 (x)$', color='b', fontsize=20)
ax.text(3.3, 2.6, '$p_{2,3}^1 (x)$', color='b', fontsize=20)
ax.text(6.4, 1.85, '$p_{3,4}^1 (x)$', color='b', fontsize=20)

ax.text(3.3, 3.95, '$p_{1,3}^2 (x)$', color='r', fontsize=20)
ax.text(3.4, 1.1, '$p_{2,4}^2 (x)$', color='m', fontsize=20)

ax.text(3.3, 3.4, '$p_{1,4}^3 (x)$', color='g', fontsize=20)

ax.set_title("Interpolation Polynomials Used For Padding", fontsize=25)

plt.tight_layout()
fig.savefig("images/padding_polynomials.png")
fig.show()
```
