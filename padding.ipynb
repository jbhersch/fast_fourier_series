```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
from ffs import FFS
import seaborn as sns

sns.set_style("whitegrid")

%matplotlib inline
```

```python
x = np.linspace(0,10,101)
# y = np.sin(2*np.sin(2*np.sin(2*np.sin(x))))
y = np.sin(np.pi*x/5)-np.tan(x*2)
y = x**

fs = FFS(x, y, pad=1001)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

# ax.plot(x,y,'b',linewidth=10,alpha=0.25)
ax.plot(x, fs.evaluate(x,deriv=1), 'r')

fig.show()
```

```python
x = np.linspace(0,10,101)
# y = np.sin(2*np.sin(2*np.sin(2*np.sin(x))))
y = np.sin(2*np.sin(2*np.sin(2*np.sin(x))))
y = x**3

# fs = FFS(x, y, pad=100)
l2 = []
for pad in xrange(500):
    fs = FFS(x, y, pad=pad)
    l2.append(np.linalg.norm(y-fs.evaluate(x)))
```

```python
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)

ax.plot(range(500), l2, 'o')

fig.show()
```
