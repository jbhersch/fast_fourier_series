```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
from ffs import FFS
import seaborn as sns

sns.set_style("whitegrid")

%matplotlib inline
```

```python
x = np.linspace(-2,2,50)
y = np.exp(-x**2)
d1_y = -2*x*y
d2_y = 2*(2*x**2 - 1)*y
fs = FFS(x,y,0)
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs.evaluate(x),'r', label = 'Fourier Series Approximation')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.ylim((-0.1, 1.1))
plt.title('Fourier Series Representation of the Standard Normal Bell Curve', fontsize=25)
plt.legend(prop={'size':20})

plt.tight_layout()
# plt.savefig('images/bell_curve.png')
plt.show()
```

```python
xx = np.hstack((x, fs.xb + fs.h))


plt.figure(figsize=(15,10))
plt.plot(xx,fs.evaluate(xx),'b', label="First Period")
plt.plot(xx + fs.L,fs.evaluate(xx + fs.L),'r', label="Second Period")
plt.title("Bell Curve With Two Periods", fontsize=25)
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.xlim((-2.5,6.5))
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
plt.ylim((-0.1, 1.1))
plt.legend(loc = 'upper left', prop={'size':20})
plt.tight_layout()

# plt.savefig('images/bell_curve/bell_curve_two_periods.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,d1_y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs.evaluate(x,deriv=1),'r', label = 'Fourier Series Approximation')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
# plt.ylim((-0.1, 1.1))
plt.title('First Derivative of Standard Normal Bell Curve', fontsize=25)
plt.legend(prop={'size':20})

plt.tight_layout()
# plt.savefig('images/bell_curve_1st_derivative.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,d2_y,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs.evaluate(x,deriv=2),'r', label = 'Fourier Series Approximation')
plt.xlabel('x', fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('y', fontsize=20)
plt.yticks(fontsize=20)
# plt.ylim((-0.1, 1.1))
plt.title('Second Derivative of Standard Normal Bell Curve', fontsize=25)
plt.legend(prop={'size':20})

plt.tight_layout()
# plt.savefig('images/bell_curve_2nd_derivative.png')
plt.show()
```
