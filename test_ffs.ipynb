```python
x = np.linspace(-2,2,50)
y_bell = np.exp(-x**2)
fs_bell = FFS(x,y_bell,0)

plt.figure(figsize=(15,10))
plt.plot(x,y_bell,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs_bell.evaluate(x),'r', label = 'Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim((-0.1, 1.1))
plt.title('Fourier Series Representation of the Standard Normal Bell Curve - Wave Number = 25')
plt.legend()
plt.tight_layout()

# plt.savefig('images/bell_curve.png')
plt.show()
```

```python
def remove_noise(y, thresh):
    yt = fft(y)
    yt0 = yt[0]
    yt = yt[1:]
    yt_abs = np.absolute(yt)
    yt_max = np.max(yt_abs)
    indices = np.where(yt_abs < thresh*yt_max)[0]
    yt[indices] = np.complex(0,0)
    return ifft(np.hstack((yt0, yt)))
```

```python
num_points = 100

xa = -2
xb = 2
L = xb - xa

x = np.linspace(xa,xb,num_points)
y = np.exp(-x**2)
omega = 2*np.pi/L
wave_number = 10
amplitude = 1
noise = amplitude*np.sin(wave_number*omega*(x-xa))
y_noise = y + noise

yt = fft(y_noise)
yt = 2*np.absolute(yt[1:len(yt)/2])/(len(y)-1)

plt.figure(figsize=(15,10))

plt.subplot(2,1,1)
plt.plot(x, y,'b',label="True Signal")
plt.plot(x, y_noise,'r',label="Noisy Signal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Signal In x-y Space")
plt.legend()

plt.subplot(2,1,2)
plt.stem(range(1,len(yt)+1), np.abs(yt))
plt.xticks(range(1,len(yt)+1))
plt.xlabel("Harmonic Wave Number")
plt.ylabel("Wave Amplitude")
plt.title("Signal In Complex Space")

plt.tight_layout()
plt.show()
```

```python
num_points = 100

x = np.linspace(-2,2,num_points)
amplitude = 0.2
noise = amplitude*np.random.uniform(low = -1, high = 1, size=num_points)
y = np.exp(-x**2)
y_noise = y + noise

thresh = 0.125
fs = FFS(x,y_noise,thresh = thresh)

yt = fft(y_noise)
yt = 2*np.absolute(yt[1:len(yt)/2])/(len(y)-1)

plt.figure(figsize=(15,10))

plt.subplot(2,1,1)
plt.plot(x, y,'b',label="True Signal")
plt.plot(x, y_noise,'r',label="Noisy Signal")
plt.plot(x, fs.y, 'g', label="Signal With Noise Removal")
plt.legend()

plt.subplot(2,1,2)
plt.stem(range(1,len(yt)+1), np.abs(yt))
plt.xticks(range(1,len(yt)+1))
plt.xlabel("Harmonic Wave Number")
plt.ylabel("Wave Amplitude")
plt.title("Signal In Complex Space")
plt.axhline(y=thresh*np.max(yt), color='r', linestyle='-',label="Cutoff Threshold")
plt.legend()

plt.tight_layout()
plt.show()
```
