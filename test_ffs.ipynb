```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from ffs_alpha import FFS
import seaborn as sns
sns.set_style("whitegrid")

%matplotlib inline
```

```python
x = np.linspace(-2,2,50)
y_bell = np.exp(-x**2)
fs_bell = FFS(x,y_bell,0)

plt.figure(figsize=(15,10))
plt.plot(x,y_bell,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs_bell.evaluate(x),'r', label = 'Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim((-0.1, 1.1))
plt.title('Fourier Representation of the Standard Normal Bell Curve - Wave Number = 24')
plt.legend()
plt.tight_layout()

plt.savefig('images/bell_curve.png')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(x,y_bell,'b',linewidth=10, alpha = 0.25, label = 'Input Function')
plt.plot(x,fs_bell.evaluate(x, 1),'r', label = 'Fourier Series Approximation')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim((-0.1, 1.1))
plt.title('Fourier Representation of the Standard Normal Bell Curve - Wave Number = 1')
plt.legend()
plt.tight_layout()

plt.savefig('images/bell_curve_1.png')
plt.show()
```
